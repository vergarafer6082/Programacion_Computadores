# INTRODUCCIÓN

Crear una **Tabla de Contenido (TOC)** en Google Colab es fundamental para mantener tus cuadernos organizados, especialmente cuando el análisis de datos o el código se vuelven extensos. A diferencia de un documento de texto tradicional, en Colab tienes dos formas principales de navegar: una automática y otra personalizada mediante enlaces internos.

Aquí tienes una breve guía de cómo lograrlo:

---

### 1. El Panel Lateral Automático

Google Colab genera automáticamente una tabla de contenido basada en las **celdas de texto** que utilizan encabezados de Markdown. Solo necesitas hacer clic en el icono de la "lista" (tres líneas horizontales) en la barra lateral izquierda.

### 2. Creación Manual con Markdown

Si prefieres tener una tabla de contenido directamente en una celda al principio de tu cuaderno para que sea lo primero que vea el usuario, puedes usar **enlaces internos**.

La sintaxis sigue esta lógica:

1. **Define el encabezado:** Usa `#` para secciones principales o `##` para subsecciones.
2. **Crea el enlace:** Usa la estructura `[Nombre de la Sección](#nombre-de-la-sección)`.

> **Nota importante:** Colab convierte automáticamente los espacios en guiones (`-`) y todo a minúsculas para los enlaces internos.

#### Ejemplo de código Markdown:

```markdown
# Índice
1. [Introducción](#introducción)
2. [Carga de Datos](#carga-de-datos)
3. [Conclusiones](#conclusiones)

---
## Introducción
Contenido de la intro...

## Carga de Datos
Código para importar archivos...

```

### Beneficios de una buena navegación

* **Legibilidad:** Permite que otros usuarios entiendan la estructura de tu proyecto de un vistazo.
* **Eficiencia:** No pierdes tiempo haciendo scroll infinito buscando una función específica.
* **Interactividad:** Transforma tu notebook en un informe técnico profesional.

# FUNDAMENTO TEÓRICO

El fundamento teórico de la navegación en un notebook de Google Colab se basa en dos pilares tecnológicos: el lenguaje de marcado **Markdown** y el uso de **identificadores de fragmento (anclas HTML)**.

---

### Los Pilares de la Navegación

* **Jerarquía de Encabezados (Markdown):** Colab interpreta los símbolos `#` como etiquetas de encabezado HTML (`<h1>`, `<h2>`, etc.). Al renderizar la celda, el sistema asigna automáticamente un **ID único** a cada título basado en su texto.
* **Hipervínculos Internos:** La navegación funciona mediante el protocolo de enlaces de fragmento. Cuando creas un enlace tipo `[Texto](#sección)`, el navegador busca en el **DOM** (la estructura interna de la página) un elemento cuyo ID coincida exactamente con el nombre después del `#`.
* **Normalización de IDs:** Para que el enlace funcione, Colab sigue una regla de transformación: convierte el texto a minúsculas, elimina caracteres especiales y sustituye los espacios por guiones.

> **En resumen:** La navegación no es más que una serie de "saltos" hacia marcadores invisibles que Colab genera automáticamente cada vez que escribes un título con `#`.
